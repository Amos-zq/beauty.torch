```lua
require 'image'
t = require 'transforms_beauty'
```

```lua
img = image.load('/home/eightbit/Desktop/download.jpg')
```

```lua
itorch.image(img)
```

```lua
meanstd = {
   mean = { 0.485, 0.456, 0.406 },
   std = { 0.229, 0.224, 0.225 },
}
```

```lua
function preprocess()
  return t.Compose{
     t.Scale(256),
     t.RandomSizedCrop(224),
    -- t.ColorJitter({
    --   brightness = 0.1,
    --   contrast = 0.1,
    --    saturation = 0.1,
    -- }),
     t.Rotation(45),
     t.ColorNormalize(meanstd),
     t.HorizontalFlip(0.5),
  }
end
```

```lua
transform = preprocess()
```

```lua
img2 = transform(img)
itorch.image(img2)
```
